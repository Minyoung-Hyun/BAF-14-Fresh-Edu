- <p style="font-weight:bold">이슬점 온도와 기온은 0.91, 굉장히 강한 양의 상관관계를 가짐.</p>

# Functioning Day (작동일? 여부)
- 'Yes':작동일, 'No':작동하지 않은 시간으로 구성
- Functioning Day가 아닌 시간은 자전거를 사용할 수 없으므로 자전거 대여횟수가 0.
- 작동일의 자전거대여횟수와 원래데이터의 자전거대여횟수 비교  
  작동하지 않은=대여횟수가 0인 경우를 지움으로써 작동한 날의 수치를 더 정확히 파악가능  

```python
data['Functioning Day'].unique()
```

```python
data['Functioning Day'].value_counts()
```

```python
sns.countplot(x='Functioning Day',data=data)
plt.title('Functioning Day barplot',fontsize=20)
```

```python
#Functioning Day(작동일)이 No(아닌 날)의 Rented Bike Count(자전거대여횟수)
print(data[data['Functioning Day']=='No']['Rented Bike Count'])
#작동하지 않으므로 자전거 대여횟수가 0암.
```

```python
#Functioning Day의 데이터만 모음
function_count=data[data['Functioning Day']=='Yes']['Rented Bike Count']
print(function_count)
```

```python
#Functioning Day(작동)힌 날의 자전거 대여횟수
sns.boxplot(function_count)
plt.title("Barplot of Functioning Day")
```

```python
#작동일의 자전거대여횟수(unction_count)와 원래 데이터의 자전거대여횟수 비교
sns.distplot(function_count)
sns.distplot(data['Rented Bike Count'], color='red')
```
